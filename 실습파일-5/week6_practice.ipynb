```python
# matplotlib를 이용하여 데이터 시각화

import pandas as pd
df = pd.read_csv('mpg.csv')
print(df)
print()
```

```python

import matplotlib.pyplot as plt
df1 = df[['mpg', 'acceleration']].iloc[:20]
print(df1)
print()
```

```python
# 그래프의 스타일을 ggplot으로 설정하여, 그래프의 색상과 배경이 깔끔하고 보기 좋게 설정
plt.style.use('ggplot')

# kind='bar' : 막대 그래프로 지정, 
# figsize=(14, 5) : 그래프의 크기를 가로 14인치, 세로 5인치로 설정
# width=0.7: 막대의 너비를 조절, 0.7은 기본값보다 약간 더 얇은 막대
# color : 두 가지 색상으로 막대를 색칠
df1.plot(kind='bar', figsize=(14, 5), width=0.7,
          color=['skyblue', 'blue'])

# 그래프 제목, 제목의 글씨 크기 지정
plt.title('MPG & ACCELERATION', size=30)
# x축의 레이블을 "Index"로 설정하고, 글씨 크기를 20으로 지정 -> x축에는 데이터 프레임의 인덱스가 표시
plt.xlabel('Index', size=20)
# 그래프에 범례(legend)를 표시합, loc='best'는 범례가 그래프 내에서 가장 적절한 위치에 자동으로 배치되도록 한다.
plt.legend(loc='best', fontsize=15)
# 그래프를 화면에 표시
plt.show()
```

```python
# y축 하나 x축 하나인 막대 그래프

import matplotlib.pyplot as plt
plt.style.use('ggplot')

# df 데이터프레임에서 origin 열을 선택, value_counts() 함수는 이 열에 있는 각 값의 빈도수를 계산
df3 = df['origin'].value_counts()
print(df3)
print()

# 막대그래프, 크기는 가로 7인치 세로 5인치, 막대의 너비는 0.7로 설정, 색깔은 skyblue
df3.plot(kind='bar', figsize=(7, 5), width=0.7,
          color='skyblue')

plt.title('ORIGIN', size=30)
plt.xlabel('Origin', size=20)
plt.legend(loc='best', fontsize=15)

plt.show()
```

```python
# y축의 변수를 2개 그리는 막대 그래프

import matplotlib.pyplot as plt
df1 = df[['mpg', 'weight']].iloc[:20]
print(df1)
print()
plt.style.use('ggplot')

# df1의 mpg컬럼을 막대 그래프로 그림 -> y1
ax = df1['mpg'].plot(kind='bar', figsize=(14, 5), width=0.7,
          color = 'skyblue', label = 'mpg')

# twinx()는 동일한 x축을 공유하는 두 번째 y축을 추가
# 두 개의 y축이 서로 다른 데이터를 표현할 수 있게 한다.
ax1 = ax.twinx()
# weight 컬럼을 y2로, x축도 지정해서 그래프를 그림
# df1.index : x축으로는 df1의 인덱스 사용
# marker='o': 선의 데이터 포인트(weight)를 원형(marker)으로 표시, 크기는 8 (markersize)
# label='weight': 범례에 표시될 라벨을 "weight"로 설정
ax1.plot(df1.index, df1['weight'], marker='o', markersize=8, label='weight')


ax.set_title('MPG & WEIGHY', size=30, color='brown', weight='bold')
ax.set_ylabel('mpg', size=20, color='blue')
ax1.set_ylabel('weight', size=20, color='blue')
ax.legend(loc='upper left', fontsize=15)
ax1.legend(loc='upper right', fontsize=15)

plt.show()
```

```python
# 히스토그램으로 그리기

import matplotlib.pyplot as plt
plt.style.use('ggplot')

df4 = df['mpg']

df4.plot(kind='hist', figsize=(7, 5), bins=10,
          color='skyblue')

plt.title('HISTOGRAM', size=30)
plt.xlabel('Index', size=20)
plt.legend(loc='best', fontsize=15)

plt.show()
```

```python
# scatter, 산점도 그리기

import matplotlib.pyplot as plt
df1 = df[['mpg', 'weight']]
plt.style.use('ggplot')

df1.plot(kind='scatter', x='weight', y='mpg',
         figsize=(14, 10), color = 'skyblue')

plt.title('MPG & WEIGHY', size=30, color='brown', weight='bold')
plt.ylabel('mpg', size=20, color='blue')
plt.xlabel('weight', size=20, color='blue')
plt.show()
```

```python
# scatter, 산점도 그리기

import matplotlib.pyplot as plt
df1 = df[['mpg', 'weight', 'acceleration', 'displacement']]
plt.style.use('ggplot')

# 여기서는 weight와 mpg 간의 관계를 시각화하면서 acceleration과 displacement 데이터를 추가 정보로 사용
# c=df1['acceleration']: 점의 색상은 acceleration 값에 따라 결정, 즉, 가속도 값에 따라 각 점의 색상이 변합
# s=df1['displacement']: 점의 크기는 displacement 값에 따라 결정, 배기량이 큰 데이터일수록 점의 크기가 커집
df1[['mpg', 'weight']].plot(kind='scatter', x='weight', y='mpg',
         figsize=(14, 10), c = df1['acceleration'], s = df1['displacement'])

plt.title('MPG & WEIGHY', size=30, color='brown', weight='bold')
plt.ylabel('mpg', size=20, color='blue')
plt.xlabel('weight', size=20, color='blue')
plt.show()
```

```python
# 파이차트 그리기

import matplotlib.pyplot as plt
plt.style.use('ggplot')

# df3 : df 프레임에서 'origin'이라는 컬럼을 뽑아 각 value의 개수를 나타내는 데이터프레임
df3 = df['origin'].value_counts()
# figsize=(7, 5): 그래프의 크기를 가로 7인치, 세로 5인치로 설정
# startangle=10: 파이 차트가 그려질 때의 시작 각도를 10도로 설정하여, 차트의 회전 방향을 약간 조정
# autopct='%.2f%%': 파이 차트에서 각 조각의 비율을 퍼센트로 표시, 소수점 둘째 자리까지 나타냄
# colors=['blue', 'skyblue', 'green']: 파이 차트의 각 조각에 적용할 색상을 지정
df3.plot(kind='pie', figsize=(7, 5), startangle=10,
         autopct='%.2f%%',
          colors=['blue', 'skyblue', 'green'])

# 파이 차트에는 Y축이 의미가 없기 때문에 Y축 레이블을 빈 문자열로 설정하여 Y축 레이블을 제거
plt.ylabel("")
# 그래프의 제목을 "Pie Chart - Origin"으로 설정하고, 글씨 크기를 15로 지정
plt.title('Pie Chart - Origin', size=15)
# 범례(legend)를 그래프의 오른쪽 상단(upper right)에 표시하고, 글씨 크기를 8로 지정하여 범례를 작게 나타냄
plt.legend(loc='upper right', fontsize=8)

plt.show()
```

```python
# 선그래프 그리기

import pandas as pd
import matplotlib.pyplot as plt

# mpg.csv 파일을 읽어와 데이터프레임 df로 저장합니다. df에는 mpg.csv 파일의 모든 데이터가 저장
df = pd.read_csv('mpg.csv')
plt.style.use('ggplot') 

plt.figure(figsize=(14, 5))
# X축 눈금(ticks)의 글자 크기를 10으로 설정하고, 눈금 레이블을 수직으로 회전, 이로 인해 X축의 레이블이 수직으로 표시
plt.xticks(size=10, rotation='vertical')
# df 데이터프레임에서 인덱스를 X축으로, mpg 값을 Y축으로 하여 선 그래프를 그림
plt.plot(df.index, df['mpg'], marker='o', markersize=8)

plt.title('auto-mpg', size=30) 
plt.xlabel('mpg', size=20)
plt.ylabel('index', size=20)
# 범례(legend)를 추가하여 그래프에서 mpg 데이터가 무엇을 나타내는지 설명
# loc='best'는 범례가 그래프 내에서 가장 적절한 위치에 자동으로 배치하며, 글씨 크기는 10으로 설정
plt.legend(labels=['mpg'], loc='best', fontsize=10)

plt.show()
```

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot') 

# 새로운 figure(그림)을 생성하고, 크기를 가로 14인치, 세로 10인치로 설정, 이 figure는 여러 개의 서브플롯을 포함할 수 있는 컨테이너 역할
fig = plt.figure(figsize=(14, 10))   
# 첫 번째 서브플롯을 만듦, add_subplot(2, 1, 1)는 2행 1열짜리 서브플롯중 첫번째 지정(위)
ax1 = fig.add_subplot(2, 1, 1)
# 두 번째 서브플롯을 만듦,  add_subplot(2, 1, 2)는 2행 1열짜리 서브플롯중 두번째 지정(아래)
ax2 = fig.add_subplot(2, 1, 2)

# 각 서브플롯에 선 그래프 그리기
ax1.plot(df.index, df['mpg'], marker='o', markersize=8, label = 'mpg')
ax2.plot(df.index, df['weight'], marker='o', markerfacecolor='green', markersize=8, 
         color='olive', linewidth=2, label='weight')
ax1.legend(loc='best')
ax2.legend(loc='best')

# 첫 번째 서브플롯의 Y축 값의 범위를 0에서 50까지로 설정, 이는 mpg 데이터가 이 범위 내에서만 나타나도록 한다.
ax1.set_ylim(0, 50)
# 두 번째 서브플롯의 Y축 값의 범위를 1000에서 6000까지로 설정
ax2.set_ylim(1000, 6000)

# 두 서브플롯의 X축 눈금 레이블을 설정, X축 레이블을 df.index로 지정하고, 90도 회전하여 세로 방향으로 표시되도록 설정
ax1.set_xticklabels(df.index, rotation=90)
ax2.set_xticklabels(df.index, rotation=90)

plt.show()
```

```python
import matplotlib.pyplot as plt

# figure 생성 
fig = plt.figure(figsize=(14, 5))   
# 1행 1열짜리 서브 플롯 지정
ax = fig.add_subplot(1, 1, 1)

# 하나의 서브플롯에 두 개의 그래프 그리기
# linewidth=2: 선의 두께를 2로 설정하여 굵게 표시
ax.plot(df.index, df['mpg'], marker='o', markerfacecolor='green', 
        markersize=10, color='olive', linewidth=2, label='mpg')
ax.plot(df.index, df['acceleration'], marker='o', markerfacecolor='blue', 
        markersize=10, color='skyblue', linewidth=2, label='acceleration')

ax.legend(loc='best')
ax.set_title('MPG & ACCELERATION', size=20)
# X축의 레이블을 "index"로 설정하고, 글씨 크기를 12로 지정
ax.set_xlabel('index', size=12) 
# X축의 눈금 레이블을 df.index로 설정하고, 레이블을 0도로 설정하여 가로 방향으로 그대로 표시되도록 한다.
ax.set_xticklabels(df.index, rotation=0)

# X축과 Y축의 눈금 레이블 크기를 각각 12로 설정
ax.tick_params(axis="x", labelsize=12)
ax.tick_params(axis="y", labelsize=12)

plt.show()
```

```python
# 영역 그래프(area chart) 그리기

import matplotlib.pyplot as plt
df1 = df[['mpg', 'acceleration']]
plt.style.use('ggplot')

# kind='area': 그래프의 종류를 영역 그래프(면적 그래프)로 설정
# stacked=False: 각 데이터의 영역을 서로 겹치지 않도록 설정
# alpha=0.2: 영역의 투명도를 0.2로 설정, 이 값은 0(완전 투명)에서 1(완전 불투명) 사이의 값으로, 영역이 약간 투명하게 보이도록 한다.
# figsize=(14, 5): 그래프의 크기를 가로 14인치, 세로 5인치로 설정
df1.plot(kind='area', stacked=False, alpha=0.2, figsize=(14, 5))

plt.title('MPG & ACCELERATION', size=30, color='brown', weight='bold')
plt.xlabel('Index', size=20)
plt.legend(loc='best', fontsize=15)

plt.show()
```

```python
# 영역 그래프(area chart) 그리기, ax라는 새 변수에 담아서 설정할 수도 있음 

import matplotlib.pyplot as plt
df1 = df[['mpg', 'acceleration']]
plt.style.use('ggplot')

# stacked=True: 두 데이터(mpg와 acceleration)의 영역을 겹쳐서 쌓이도록 설정. 즉, 영역들이 중첩되어 보여진다.
ax = df1.plot(kind='area', stacked=True, alpha=0.2, figsize=(20, 10))

ax.set_title('MPG & ACCELERATION', size=30, color='brown', weight='bold')
ax.set_xlabel('Index', size=20, color='blue')
ax.legend(loc='best', fontsize=15)

plt.show()
```
